## Time Resampling

```python
df_tesla.head()
```

```python
df_tesla.resample(rule='A')
```

```python
df_tesla.resample(rule='A').min()
```

```python
df_tesla.resample(rule='A').max()
```

```python
#rule A year,frequency
df_tesla.resample(rule='A').max()['Open'].plot()
```

```python
#quaterly
df_tesla.resample(rule='QS').max()['High'].plot()
```

```python
## bussiness and frequency
df_tesla.resample(rule='BA').max()
```

```python
df_tesla.resample(rule='BQS').max()
```

```python
#plotting
df_tesla.resample(rule='BA').mean().plot(kind='bar')
```

```python
df_tesla.resample(rule='M').mean().plot(kind='bar',figsize=(15,6))
```

```python
df_tesla.resample(rule='M').max().plot(kind='bar',figsize=(15,6))
```
